```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()
```

```python
xs
```

```python
params ={'n_neighbors':st.randint(2,50),
        'p':st.randint(1,5)}
```

```python
rsearch = RandomizedSearchCV(knc,param_distributions=params,cv=3,n_iter=100,scoring='roc_auc',n_jobs=-1)
rsearch.fit(xs,y)
```

```python
rsearch.best_params_
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
# fit_transform for x_train and transform for x_test
```

```python
xtrain_std = ss.fit_transform(x_train)
xtest_std = ss.transform(x_test)
```

```python
# Build KNN
```

```python
knc = KNeighborsClassifier(n_neighbors=11,p=2)
knc.fit(xtrain_std,y_train)
```

```python
ytrain_pred = knc.predict(xtrain_std)
ytrain_prob = knc.predict_proba(xtrain_std)[:,1]
```

```python
ytest_pred = knc.predict(xtest_std)
ytest_prob = knc.predict_proba(xtest_std)[:,1]
```

```python
roc_auc_score(y_train,ytrain_prob)
```

```python
roc_auc_score(y_test,ytest_prob)
```

# Stacking Algorithm to improve results and find best good algorithm using classification

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
```

```python
knc = KNeighborsClassifier(n_neighbors=11,p=2)
lr = LogisticRegression(solver='liblinear')
rfc = RandomForestClassifier(**rsearch.best_params_)

```

```python
clf = VotingClassifier(estimators=[('le',lr),('rfc',rfc),('knc',knc)],voting='soft')
# we can also give weights
#clf = VotingClassifier(estimators=[('le',lr),('rfc',rfc),('knc',knc)],voting='soft',weights=[2,3,1])
```

```python
clf.fit(x_train,y_train)
```

```python
ytrain_pred = clf.predict(x_train)
ytrain_prob = clf.predict_proba(x_train)[:,1]
```

```python
accuracy_score(y_train,ytrain_pred)
```

```python
roc_auc_score(y_train,ytrain_prob)
```

```python
ytest_pred = clf.predict(x_test)
ytest_prob = clf.predict_proba(x_test)[:,1]
```

```python
accuracy_score(y_test,ytest_pred)
```

```python
roc_auc_score(y_test,ytest_prob)
```
