```python
import pandas as pd
```

```python
df = pd.read_csv('SMSSpamCollection.csv')
```

```python
df
```

```python
df['Class'].replace({'ham':0,'spam':1},inplace=True)
```

```python
df
```

```python
df['Class'].value_counts()
```

```python
x = df['sms']
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
# we need to convert each word as a feature
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
```

```python
x_traint = vect.fit_transform(x_train)
x_testt = vect.transform(x_test)
```

```python
pd.DataFrame(x_traint)
```

```python
x_traint.shape
```

```python
x_traint
```

```python
vect.get_feature_names()
```

```python
from sklearn.naive_bayes import BernoulliNB
nb = BernoulliNB()
```

```python
nb.fit(x_traint,y_train)
```

```python
ytrain_pred = nb.predict(x_traint)
ytrain_prob = nb.predict_proba(x_traint)[:,1]
```

```python
ytest_pred = nb.predict(x_testt)
ytest_prob = nb.predict_proba(x_testt)[:,1]
```

```python
from sklearn.metrics import accuracy_score,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_train,ytrain_pred)
```

```python
accuracy_score(y_test,ytest_pred)
```

```python
fpr,tpr,tresh=roc_curve(y_train,ytrain_prob)
```

```python
tresh[0] = tresh[0]-1
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(fpr,tpr)
plt.plot(fpr,fpr)
plt.plot(fpr,tresh)
```

```python
ds = pd.DataFrame({'fpr':fpr,'tpr':tpr,'tresh':tresh})
```
